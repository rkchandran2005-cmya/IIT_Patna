# Ecommerce Sales Analysis

## Problem Statement
Analyze sales data to identify performance across regions, categories, and salespersons using grouping and aggregation techniques in Pandas.

```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate data
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

df.head(10)
print("Dataset shape:", df.shape)
```

## ✅ Task 1: Basic Grouping and Single Aggregations

### Approach:
Use `groupby()` with aggregation functions (`sum`, `count`, `mean`), apply `reset_index()` for clean output, and use sorting to identify top performers.

```python
# 1. Total sales per region + sorting
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales = region_sales.sort_values(by='sales_amount', ascending=False)
print("Total Sales by Region:")
print(region_sales)
```

```python
# 2. Transaction count per category
category_count = df.groupby('category')['transaction_id'].count().reset_index()
print("\nTransactions per Category:")
print(category_count)
```

```python
# 3. Average sales per salesperson
avg_salesperson = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAverage Sales per Salesperson:")
print(avg_salesperson)
```

### Explanation:
- `groupby()` is used to group data based on columns.
- `sum()` calculates total sales.
- `count()` counts number of transactions.
- `mean()` calculates average sales.
- `reset_index()` converts grouped results into a proper DataFrame.
- `sort_values()` is used to identify the top-performing region.

## ✅ Task 2: Multi-Column Grouping and Multiple Aggregations

### Approach:
Use multiple columns in `groupby()` and apply `agg()` to compute multiple metrics simultaneously.

```python
# 1. Region + Category sales
region_category_sales = df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()
print("Region & Category Sales:")
print(region_category_sales)
```

```python
# 2. Salesperson performance
salesperson_stats = df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
salesperson_stats = salesperson_stats.sort_values(by='sum', ascending=False)
print("\nSalesperson Performance:")
print(salesperson_stats)
```

```python
# 3. Top category
category_sales = df.groupby('category')['sales_amount'].sum()
top_category = category_sales.idxmax()
print("\nTop Category:", top_category)
```

### Explanation:
- Multiple columns can be used in `groupby()` for detailed analysis.
- `agg()` allows multiple aggregations at once.
- Sorting helps identify top-performing salesperson.
- `idxmax()` returns the category with highest sales.

## ✅ Task 3: Custom Aggregation and Complete Sales Report

### Approach:
Use custom functions and dictionary-based aggregation for advanced analysis.

```python
# Custom function
def sales_range(x):
    return x.max() - x.min()

# Region analysis
region_analysis = df.groupby('region')['sales_amount'].agg(
    ['sum', 'mean', 'min', 'max', sales_range]
).reset_index()

print("Region Analysis:")
print(region_analysis)
```

```python
# Final summary report
final_report = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

print("\nFinal Report:")
print(final_report)
```

### Multi-Level Column Explanation:
- When multiple aggregations are applied to the same column, Pandas creates multi-level (hierarchical) column names.
- Example: ('sales_amount', 'sum'), ('sales_amount', 'mean')

### Difference:
- Single aggregation → simple column
- Multiple aggregation → multi-level columns

### Conclusion:
- Groupby helps in summarizing data
- Aggregations provide insights into performance
- Custom functions add flexibility
- Sorting and indexing help identify top performers
