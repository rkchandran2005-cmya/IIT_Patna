## Mini student grade analyzer

```python
def process_scores(students):
    averages = {}

    for name, scores in students.items():
        avg = round(sum(scores) / len(scores), 2)
        averages[name] = avg

    return averages


def classify_grades(averages):
    A = 90
    B = 80
    C = 70
    D = 60

    result = {}

    for name, avg in averages.items():

        if avg >= A:
            grade = "A"
        elif avg >= B:
            grade = "B"
        elif avg >= C:
            grade = "C"
        elif avg >= D:
            grade = "D"
        else:
            grade = "F"

        result[name] = (avg, grade)
    return result


def generate_report(classified, passing_avg=70):
    print("===== Student Grade Report =====")
    total = len(classified)
    passed = 0
    failed = 0
    for name, (avg, grade) in classified.items():
        if avg >= passing_avg:
            status = "PASS"
            passed += 1
        else:
            status = "FAIL"
            failed += 1

        print(f"{name:<10} | Avg: {avg:.2f} | Grade: {grade} | Status: {status}")
    print("================================")
    print(f"Total Students : {total}")
    print(f"Passed         : {passed}")
    print(f"Failed         : {failed}")

    return passed

students = {
    "Alice": [80, 90, 85, 90],
    "Bob": [60, 65, 55, 70],
    "Clara": [95, 98, 96, 96]
}

averages = process_scores(students)
classified = classify_grades(averages)
passed_students = generate_report(classified)

print("\nReturned Passed Count:", passed_students)
```
