### This step imports the pandas library to work with DataFrames in Python.

```python
import pandas as pd
```

### This step creates a pandas DataFrame containing student information such as ID, Name, Age, Score, and Grade.

```python
data = {
    "ID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","David","Emma","Frank","George","Hannah","Ivan","Julia"],
    "Age": [20,21,22,23,19,21,23,19,22,20],
    "Score": [85,78,92,88,76,90,70,95,60,80],
    "Grade": ["B","C","A","B","C","A","B","A","C","B"]
}

df = pd.DataFrame(data)

df
```

### This step saves the DataFrame to a CSV file for later use.

```python
df.to_csv("students.csv", index=False)
```

### This step reloads the saved CSV file into a new DataFrame.

df = pd.read_csv("students.csv")
